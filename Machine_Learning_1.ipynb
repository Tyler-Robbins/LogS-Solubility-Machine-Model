# **First Machine-Learning Project**

# Loading Dataset

```python
import pandas as pd
dataFrame = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv')
dataFrame
```

# **Data Preperation**

## Data Seperation as X and Y

```python
yValue = dataFrame['logS']
yValue
```

```python
xValue = dataFrame.drop('logS', axis=1)
xValue
```

## Data Splitting

```python
from sklearn.model_selection import train_test_split

x_Train, x_Test, y_Train, y_Test = train_test_split(xValue, yValue, test_size=0.2, random_state=100)

x_Train
x_Test
```

x_Train and x_Test splits the data set into 80% and 20% forms of the data in order to use the train data to train the model and 20% to test the data. https://towardsdatascience.com/how-to-build-a-machine-learning-model-439ab8fb3fb1

# **Model Building**



## **Linear Regression**

### **Training the model**

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_Train, y_Train)
```

### **Applying the model to make a prediction**

```python
y_lr_train_pred = lr.predict(x_Train)
y_lr_test_pred = lr.predict(x_Test)

print(y_lr_train_pred)
```

```python
print(y_lr_test_pred)
```

### **Evaluate Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_Train, y_lr_train_pred)
lr_train_r2 = r2_score(y_Train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_Test, y_lr_test_pred)
lr_test_r2 = r2_score(y_Test, y_lr_test_pred)

```

```python
print('Linear Regression Mean Squared Error (Train): ', lr_train_mse)
print('Linear Regression Regression Squared (Train): ', lr_train_r2)
print('Linear Regression Mean Squared Error (Test): ', lr_test_mse)
print('Linear Regression Regression Squared (Test): ', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear Regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

## Random Forest

### **Training the model**

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(x_Train, y_Train)
```

### **Applying the model to make a prediction**

```python
y_rf_train_pred = rf.predict(x_Train)
y_rf_test_pred = rf.predict(x_Test)

print(y_rf_train_pred)
```

```python
print(y_rf_test_pred)
```

### **Evaluate Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_Train, y_rf_train_pred)
rf_train_r2 = r2_score(y_Train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_Test, y_rf_test_pred)
rf_test_r2 = r2_score(y_Test, y_rf_test_pred)

```

```python
print('Random Forest Mean Squared Error (Train): ', rf_train_mse)
print('Random Forest Regression Squared (Train): ', rf_train_r2)
print('Random Forest Mean Squared Error (Test): ', rf_test_mse)
print('Random Forest Regression Squared (Test): ', rf_test_r2)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
rf_results
```

## Model Comparison

```python
df_models = pd.concat([lr_results, rf_results], axis=0).reset_index(drop=True)
df_models
```

# Data Visualization of Results

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(5,5))
plt.scatter(x=y_Train, y=y_lr_train_pred, c='green',alpha=0.3)

z = np.polyfit(y_Train, y_lr_train_pred, 1)
p = np.poly1d(z)

plt.plot(y_Train, p(y_Train), '#F8766D')
plt.ylabel('Predicted LogS')
plt.xlabel('Experimental LogS')
plt.title('Linear Regression Experimental LogS vs Predicted LogS (Scatter)')
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x=y_Train, y=y_rf_train_pred, c='blue',alpha=0.3)

z = np.polyfit(y_Train, y_rf_train_pred, 1)
p = np.poly1d(z)

plt.plot(y_Train, p(y_Train), '#0000FF')
plt.ylabel('Predicted LogS')
plt.xlabel('Experimental LogS')
plt.title('Random Forest Experimental LogS vs Predicted LogS (Scatter)')
```

```python
plt.hist(y_Train, bins=10, alpha=0.5, label='Predicted LogS', color='blue')
plt.hist(y_lr_train_pred, bins=10, alpha=0.5, label='Experimental LogS', color='green')

plt.xlabel('Experimental LogS')
plt.ylabel('Predicted LogS')
plt.legend(loc='upper left')

plt.title('Linear Regression LogS vs Predicted LogS (Bar Graph)')


plt.show()
```

```python
plt.hist(y_Train, bins=10, alpha=0.5, label='Predicted LogS', color='blue')
plt.hist(y_rf_train_pred, bins=10, alpha=0.5, label='Experimental LogS', color='green')

plt.xlabel('Experimental LogS')
plt.ylabel('Predicted LogS')
plt.legend(loc='upper left')

plt.title('Random Forest Experimental LogS vs Predicted LogS (Bar Graph)')

plt.show()
```
